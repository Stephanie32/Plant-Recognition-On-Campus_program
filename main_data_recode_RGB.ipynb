```python
import os
import math
from PIL import Image
import numpy as np
import h5py
```

```python
data=[]
```

```python
#rotate_time=360/rotate_angle
rotate_angle=20
rotate_time=18
```

```python
#新存檔法，把資料夾全丟進Train
label=0
for dirname in os.listdir("C:\pythonwork\leaf\\Train"):
    for imgname in os.listdir("C:\pythonwork\leaf\\Train\\" + dirname):
        img = Image.open("C:\pythonwork\leaf\\Train\\" + dirname + "\\" +imgname)
    
        #從4032*3024縮成252*189
        img = img.resize((252, 189)) 
        (img_x,img_y)=img.size

        #切成100*100
        x = (img_x-100)/2
        y = (img_y-100)/2
        w = 100
        h = 100
        ima = img.crop((x, y, x+w, y+h))
        #切成圓形
        r2 = 100 #圓直徑=100
        imb = Image.new('RGB', (r2, r2),(0))
        pima = ima.load()
        pimb = imb.load()
        r = float(r2/2) #圆心横坐标
        for i in range(r2):
            for j in range(r2):
                lx = abs(i-r+0.5) #到圆心距离的横坐标
                ly = abs(j-r+0.5) #到圆心距离的纵坐标
                l  = pow(lx,2) + pow(ly,2)
                if l <= pow(r, 2):
                    pimb[i,j] = pima[i,j]

        j=0
        for i in range(0,rotate_time): #邊旋轉邊存        
            imb_temp = imb.rotate(j)
            arr = np.asarray(imb_temp, dtype = np.float32)
            arr_flo = arr/255
            arr_flo[0,0,0]=label
            data.append(arr_flo)
            j=j+rotate_angle

        imb = imb.transpose(Image.FLIP_LEFT_RIGHT) #鏡像

        j=0
        for i in range(0,rotate_time):#邊旋轉邊存        
            imb_temp = imb.rotate(j)
            arr = np.asarray(imb_temp, dtype = np.float32)
            arr_flo = arr/255
            arr_flo[0,0,0]=label
            data.append(arr_flo)
            j=j+rotate_angle
        
        
    label += 1
```
